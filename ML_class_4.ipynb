```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/CSV FILE/student_scores.csv')
dataset.head()
```

```python
dataset.plot(x='Hours', y='Scores', style='y')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
X = dataset.iloc[:, :-1]. values
y = dataset.iloc[:, 1]. values
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size =0.2, random_state =0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression ()
regressor.fit(X_train , y_train)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
df = pd.DataFrame ({'Actual': y_test , 'Predicted': y_pred})
df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/CSV FILE/diabetes.csv')
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
x = dataset.drop ('Outcome' , axis = 1)
y = dataset['Outcome']
X_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
regressorq = LinearRegression()
regressorq.fit(X_train,y_train)
```

```python
print(regressorq.intercept_)
print(regressorq.coef_)
```

```python
y_pred = regressorq.predict(x_test)
df = pd.DataFrame ({'Actual': y_test , 'Predicted': y_pred})
df.head()
```

```python
print('Mean Squared Error:', metrics.mean_squared_error(y_test , y_pred))
```

```python
df.plot(x='Actual', y='Predicted', style='o')
plt.title('Actual vs Predicted')
plt.xlabel(' Actual Studied')
plt.ylabel('Percentage Score')
plt.show()
```
